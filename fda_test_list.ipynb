```python
# 讀取資料
filepath = 'C:/Users/Tibame_T14/Desktop/專題報告/使用資料/Recipe_in_FDA_2.csv'
df = pd.read_csv(filepath)

re_emoji = re.compile(
    "(["
    "\U00010000-\U0010ffff"
    "\U0001F1E0-\U0001F1FF"
    "\U0001F300-\U0001F5FF"
    "\U0001F600-\U0001F64F"
    "\U0001F680-\U0001F6FF"
    "\U0001F700-\U0001F77F"
    "\U0001F780-\U0001F7FF"
    "\U0001F800-\U0001F8FF"
    "\U0001F900-\U0001F9FF"
    "\U0001FA00-\U0001FA6F"
    "\U0001FA70-\U0001FAFF"
    "\U00002702-\U000027B0"
    "\\ufe0f-\\ufffd"
    "\\u10e6-\\u2764"
    "])"
)

def strip_emoji(text):
    return re_emoji.sub(r'', text)
# 將欄位讀出來去除emoji後重新賦值
for i in df.index:
    title = df.iloc[i, 0]
    recipe = df.iloc[i, 3]
    quantity = df.iloc[i, 4]
    df.iloc[i, 0] = strip_emoji(title)
    df.iloc[i, 3] = strip_emoji(recipe)
    df.iloc[i, 4] = strip_emoji(quantity)
# 插入id    
df.insert(0,'id',df.index)

# 存成csv
df.to_csv('Recipe_in_FDA_3.csv', index=False)
```
